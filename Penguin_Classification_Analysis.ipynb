```python
plt.figure(figsize=(5,5))
sns.distplot(df.culmen_length_mm)
plt.title("The Distribution of culmen_length_mm",
          loc='left',
          fontdict={'fontsize':20,'color':'orange'})
plt.grid()
plt.show()
     
```

```python
plt.figure(figsize=(5,5))
sns.distplot(df.culmen_depth_mm	)
plt.title("The Distribution of culmen_depth_mm	",
          loc='left',
          fontdict={'fontsize':20,'color':'orange'})
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.distplot(df.flipper_length_mm,hist=True,color='red')
plt.title("The Distribution of flipper_length_mm",
          loc='left',
          fontdict={'fontsize':20,'color':'orange'})
plt.grid()
plt.show()
     
```

```python
b=sns.countplot(x=df.body_mass_g, data=df)
for container in b.containers:
     b.bar_label(container)
plt.title("The Count of body_mass_g",
           loc='left', 
           fontdict= {'fontsize' :20, 'color': 'lime'})
plt.show()
```

```python
sns.barplot(x=df.sex ,y=df.body_mass_g, data=df)
plt.title("The body_mass_g vs sex",
loc='left',
fontdict= {'fontsize' :20, 'color': 'green'})
plt.show()
     
```

```python
sns.violinplot (x=df. flipper_length_mm,y=df.body_mass_g) 
plt.title("The body_mass_g vs flipper_length_mm", loc='left',
fontdict= {'fontsize' :20, 'color': 'blue'})
plt.show()
```

```python
sns.boxplot (x=df. sex,y=df. body_mass_g)
plt.title("The distribution of body_mass_g vs sex", loc='left',
fontdict= {'fontsize' :20, 'color': 'blue'})
plt.show()
```

```python
sns.pairplot(df,hue='body_mass_g')
plt.show()
     
     
```

```python
plt.figure(figsize=(15,10))
for i,col in enumerate (df.select_dtypes (exclude=['object']).columns,1):
    plt.subplot(4,3,1)
    plt.title(f"Distribution of {col,}",fontdict= {'fontsize' :20, 'color': 'indigo'} )
    sns.histplot(df[col], kde=True)
    plt.tight_layout()
    plt.plot()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(), annot=True) 
plt.show()
     
```
